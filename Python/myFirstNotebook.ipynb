Import fetch_openml and pandas

```python
from sklearn.datasets import fetch_openml
import pandas as pd
import numpy as np
```

Import Tianic Dataset

```python
titanic = fetch_openml(name="titanic", version=1, as_frame=True)
df = titanic.frame
df.head()
```

Check For Missing Values

```python
#df.info()
missing_values = df.isnull().sum()
missing_values

median_age = df['age'].median
df['age'].fillna(median_age,inplace=True)
```

Check for Duplicates and Remove

```python
df.duplicated().sum()


```

Inspect Sex for inconsistancies and standardise Values

```python
df.sex.value_counts()

df['sex'] = df.sex.str.upper()

df.sex.value_counts()
```

Detect and Handle Outliers in Age and Fare

```python
df.describe()

# OUTLIERS > 1.5 IQQ

Q1_age  = df['age'].quantile(0.25)
Q3_age  = df['age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
print(IQR_age)

Q1_fare  = df['fare'].quantile(0.25)
Q3_fare  = df['fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare
print(IQR_fare)


                 
age_outliers = df[(df['age'] < (Q1_age - 1.5*IQR_age)) | (df['age'] > (Q3_age + 1.5*IQR_age))]

fare_outliers = df[(df['fare'] < (Q1_fare - 1.5*IQR_fare)) | (df['fare'] > (Q3_fare + 1.5*IQR_fare))]              


age_outliers .head()
fare_outliers .head()


```

Create a New Family_Size feature Based on Sibsp and Parch

```python
df['family_size'] = df['sibsp'] + df['Parch'] + 1

```

Convert Embarked to numerical using label encodung

```python
df.embarked.value_counts()

df['embarkedNum'] = df.groupby('embarked',sort=False).ngroup()+1
#print(df['embarkedNum'])

from sklearn.preprocessing import LabelEncoder
label_encoder =LabelEncoder()

df['embarked_encoded'] = label_encoder.fit_transform(df['embarked'])

df['embarked_encoded'].value_counts()

#df.head()

#one hot encoding 

df['embarked_S'] = df.apply(
    lambda row: 1 if(row['embarked']=='S' ) else 0,
    axis = 1
)

df['embarked_C'] = df.apply(
    lambda row: 1 if(row['embarked']=='C' ) else 0,
    axis = 1
)

df['embarked_Q'] = df.apply(
    lambda row: 1 if(row['embarked']=='Q' ) else 0,
    axis = 1
)

df.head()
```
